```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# G(s)H(s) = (s + 3) / (s^3 + 2s^2 + 3s + 4)
numerator = [1, 3]  # s + 3 的系数
denominator = [1, 2, 3, 4]  # s^3 + 2s^2 + 3s + 4 的系数

# 创建传递函数
sys_open_loop = ct.TransferFunction(numerator, denominator)

ct.root_locus(sys_open_loop)
plt.title("root")
plt.xlabel("r")
plt.ylabel("j")
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# 极点
poles = [-1, -2, -5]

# 构建分子和分母的多项式系数
numerator = [1]  # 分子系数为常数1
denominator = np.poly(poles)  # 根据极点构建分母多项式系数

# 创建传递函数
sys_from_poles = ct.TransferFunction(numerator, denominator)

# 绘制根轨迹图
ct.root_locus(sys_from_poles)
plt.title("root")
plt.xlabel("实部")
plt.ylabel("虚部")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# 设置参数
Lx = 10
Ly = 10
k = 10
Nx = 20
Ny = 20
dx = Lx / Nx
dy = Ly / Ny

# 初始化温度场
T = np.zeros((Nx+1, Ny+1))

# 设置边界条件
T[:, 0] = 100
T[0, :] = 100
T[:, -1] = 0
T[-1, :] = 0

# 迭代求解
for _ in range(10000):
    T_old = T.copy()
    for i in range(1, Nx):
        for j in range(1, Ny):
            T[i, j] = 0.25 * (T_old[i-1, j] + T_old[i+1, j] + T_old[i, j-1] + T_old[i, j+1])

# 可视化结果
x = np.linspace(0, Lx, Nx+1)
y = np.linspace(0, Ly, Ny+1)
X, Y = np.meshgrid(x, y)

fig = go.Figure(data=[go.Surface(x=X, y=Y, z=T.T)])
fig.update_layout(scene=dict(xaxis_title='x (m)', yaxis_title='y (m)', zaxis_title='Temperature (°C)'),
                  title='2D Steady State Heat Conduction')
fig.show()
```
