```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# G(s)H(s) = (s + 3) / (s^3 + 2s^2 + 3s + 4)
numerator = [1, 3]  # s + 3 
denominator = [1, 2, 3, 4]  # s^3 + 2s^2 + 3s + 4 

# 
sys_open_loop = ct.TransferFunction(numerator, denominator)

ct.root_locus(sys_open_loop)
plt.title("root")
plt.xlabel("r")
plt.ylabel("j")
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ct

# 
poles = [-1, -2, -5]

# 
numerator = [1]  # 1
denominator = np.poly(poles)  # 

# 
sys_from_poles = ct.TransferFunction(numerator, denominator)

# 
ct.root_locus(sys_from_poles)
plt.title("root")
plt.xlabel("")
plt.ylabel("")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# 
Lx = 10
Ly = 10
k = 10
Nx = 20
Ny = 20
dx = Lx / Nx
dy = Ly / Ny

# 
T = np.zeros((Nx+1, Ny+1))

# 
T[:, 0] = 100
T[0, :] = 100
T[:, -1] = 0
T[-1, :] = 0

# 
for _ in range(10000):
    T_old = T.copy()
    for i in range(1, Nx):
        for j in range(1, Ny):
            T[i, j] = 0.25 * (T_old[i-1, j] + T_old[i+1, j] + T_old[i, j-1] + T_old[i, j+1])

# 
x = np.linspace(0, Lx, Nx+1)
y = np.linspace(0, Ly, Ny+1)
X, Y = np.meshgrid(x, y)

fig = go.Figure(data=[go.Surface(x=X, y=Y, z=T.T)])
fig.update_layout(scene=dict(xaxis_title='x (m)', yaxis_title='y (m)', zaxis_title='Temperature (C)'),
                  title='2D Steady State Heat Conduction')
fig.show()
```
